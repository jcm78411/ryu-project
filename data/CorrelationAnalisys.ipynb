```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ruta_csv = 'datos_red.csv'
df = pd.read_csv(ruta_csv)
```

```python
print("Primeras filas del dataset:")
print(df.head())
```

```python
correlacion = df.corr(numeric_only=True)

print("\nMatriz de correlación:")
print(correlacion)
```

```python
sns.set(style='white')

plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlacion, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=.5)
plt.title('Matriz de Correlación')
plt.show()

```

```python
sns.pairplot(df)
plt.show()
```
